```python
from pandas_datareader import data
import yfinance as yf
from datetime import datetime as dt
from bokeh.plotting import figure, show, output_file
```

```python
start = dt(2023,1, 1)
end = dt(2023,1,15)
# data.DataReader(name="GOOG", data_source="yahoo",start=start,end=end)
data = yf.download("GOOG", start=start, end=end)
data
```

```python
def increase_decrease(closePrice, openPrice):
    if closePrice > openPrice:
        value = "Increase"
    elif openPrice > closePrice:
        value = "Decrease"
    else:
        value = "Equal"
    return value

data['status'] = [increase_decrease(cPrice, oPrice) for cPrice, oPrice in zip(data.Close, data.Open)]
```

```python
data
```

```python
data["Middle"] = (data.Open+data.Close)/2
data["Height"] = abs(data.Open-data.Close)
```

```python
# Create the figure
p = figure(x_axis_type='datetime', width=1200,
           height=300, title='Candlestick Chart')

hours_12 = 12*3600*1000
# p.rect(data.index[data.Status=="Increase"], )

# Add the candlestick glyph
p.segment(x0='Date', y0='Low', x1='Date',
          y1='High', color='Black', source=data)
p.vbar(x='Date', width=2, bottom='Open',
       top='Close', color='Green', source=data)
p.vbar(x='Date', width=2, bottom='Close',
       top='Open', color='Red', source=data)

# Add axis labels and title
p.xaxis.axis_label = 'Date'
p.yaxis.axis_label = 'Price'

# Show the plot
output_file("CandlestickChart.html")
show(p)
```
